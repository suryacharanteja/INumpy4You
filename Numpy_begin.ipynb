### multiplying each element in a 1-D sequence with the corresponding element in another sequence of the same length

```python
a = list(range(1, 5000000000))
```

```python
b = list(range(1, 5000000000))
```

```python
a
```

```python
b
```

```python
c = []
```

```python
for i in range(len(a)):
    c.append(a[i]*b[i])
```

```python
c
```
